## Data

Here I will look at a datasaet of bike share docks in San Jose, CA. data was pulled from the City of San Jose's open data [website](https://data.sanjoseca.gov/dataset/bike-share-docks1).  

```python
#Start by importing geopands library to analyze and map the data
import geopandas as gpd
```

```python
#Let's take a peak at the dataset
gpd.read_file('data/Bike_Share_Docks.geojson')
```

I see 82 rows which should represent 82 docks. I see a number of docks as well as year installed. Most improtantly, this dataset contains geographical data in the form of coordinates in the last column 'geometry'. 

Let's name this file.

```python
bikedocks = gpd.read_file('data/Bike_Share_Docks.geojson')
```

Seems like an apt name. Lets see what the size of this file is. 

```python
bikedocks.shape
```

83 rows and 13 columns. 

```python
bikedocks.info
```

```python
bikedocks.head()
```

Now that I know what's inside, let's plot! 

```python
bikedocks.plot()
```

Ok, so that's a pretty random distrobution. 

Let's look at it by year for something more informative. What's our date range? 

```python
# Let's sort by year first. 
docks_sorted = bikedocks.sort_values(by='YEAR',ascending = False)
```

```python
#Now lets look at the first and last rows. 
docks_sorted.head()
```

```python
docks_sorted.tail()
```

Ok, so we sorted the data by year and then took a look at the first and last columns to see that the first docks were installed in 2013 and the most recent ones were installed in 2019.

I wonder how many were installed in 2013, in the first year of docks? 

```python
docks_sorted['YEAR'].value_counts()
```

So 15 of the 82 stations were installed the first year. Let's see those original stations: 

```python
docks_sorted[docks_sorted.YEAR == '2013']
```

Now let's see the stations mapped by year to see how the expansion developed geographically. 

```python
docks_sorted.plot(
            figsize=(20,12),   
            column = 'YEAR',   
            legend = True,             
            legend_kwds={
               'loc': 'upper right',
               'bbox_to_anchor':(1,1)
            }               
) 
```

So now we can see that there was an intial core of stations in the middle, but newer ones have expanded the coverage area of the bike docks and thus access to the ride sharing program.
